# Diccionario de Variables del Dataset

A continuación se describe el significado de cada variable incluida en el dataset:

---

### **Year**
Año calendario del registro, abarcando desde **1960 hasta 2024**.

---

### **GDP_Current_USD**
Producto Interno Bruto (PIB) expresado en **dólares estadounidenses corrientes**.  
Representa el valor total de los bienes y servicios producidos en el país durante el año, sin ajustes por inflación.

---

### **GDP_Growth_Percent**
Tasa de crecimiento anual del PIB, expresada en **porcentaje (%)**.  
Indica cuánto creció o decreció la economía respecto al año anterior.

---

### **GDP_Per_Capita_USD**
PIB por habitante, expresado en **dólares estadounidenses**.  
Se calcula dividiendo el PIB total entre la población. Mide, de manera aproximada, el nivel promedio de ingreso/economía por persona.

---

### **Inflation_Rate_Percent**
Tasa anual de inflación al consumidor, expresada en **%**.  
Refleja el cambio promedio en los precios de bienes y servicios durante el año.

---

### **Official_Exchange_Rate**
Tipo de cambio oficial, expresado como **riales iraníes (IRR) por 1 dólar estadounidense (USD)**.

---

### **Population_Total**
Población total del país en el año correspondiente.

---

### **Population_Growth_Percent**
Tasa de crecimiento anual de la población, en **%**.  
Mide cuánto aumentó o disminuyó la población respecto al año anterior.

---

### **Exports_USD**
Valor total de las **exportaciones** del país durante el año, expresado en **USD**.

---

### **Imports_USD**
Valor total de las **importaciones**, también expresado en **USD**.

---

### **Trade_Percent_GDP**
Comercio total como porcentaje del PIB.  
Generalmente se calcula como:  
**(Exportaciones + Importaciones) / PIB × 100**

---

### **Industry_Value_Added_Percent_GDP**
Participación del sector industrial en el PIB, expresada en **%**.  
Incluye manufactura, energía, construcción y minería.

---

### **Trade_Balance**
Balanza comercial, calculada como:  
**Exportaciones − Importaciones (en USD)**  
Un valor positivo indica superávit comercial; uno negativo, déficit.

---

```python
# leer otra data
df_nuevo = pd.read_csv('fao.csv')
df_nuevo.head()
```

```python
# filtramos cuando la columna item tenga 'Agriculture' aunque sea parte de la palabra
df_agriculture = df_nuevo[df_nuevo['Item'].str.contains('Agriculture', case=False, na=False)]
df_agriculture.head()  
```

```python
# voy a traerme de la base de datos original solo desde 1970 en adelante
df_1970 = df[df['Year'] >= 1970]
df_1970.head()
```

```python
# unimos las dos bases de datos por la columna 'Year' solo me traigo el año y el value
df_agriculture_yearly = df_agriculture.groupby('Year')['Value'].sum().reset_index()
df_merged = pd.merge(df_1970, df_agriculture_yearly, on='Year', how='inner')
df_merged.head()
```

Bien ahora vamos a predecir el Value de esta nueva base de datos. 

```python
# vamos a llamar todos los modelos de regresion
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_merged.columns
```

```python
# medimos el desempeño de cada modelo
X = df_merged.drop(['Value','Unnamed: 0'], axis=1)  # variables predictoras
y = df_merged['Value']  # variable objetivo

""" Train: 1960–2010  
Valid: 2011–2017  
Test: 2018–2024 """
# hacemos el split pero teniendo en cuenta el tiempo,es decir, antes del 2000 para entrenamiento y después para test
X_train = X[X['Year'] < 2011]
y_train = y[X['Year'] < 2011]
X_valid = X[(X['Year'] >= 2011) & (X['Year'] <= 2017)]
y_valid = y[(X['Year'] >= 2011) & (X['Year'] <= 2017)]
X_test = X[X['Year'] > 2017]
y_test = y[X['Year'] > 2017]
```

```python
# medimos el desempeño de cada modelo
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=29),
    'Random Forest': RandomForestRegressor(random_state=29),
    'Support Vector Regressor': SVR()
}
```

```python
# primero vamos a normalizar y estandarizar los datos
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.drop('Year', axis=1))
X_valid_scaled = scaler.transform(X_valid.drop('Year', axis=1))
X_test_scaled = scaler.transform(X_test.drop('Year', axis=1))
```

```python
# imprimimos en una tabla el desempeño de cada modelo usando la validación solamente
# quitar los Nan de X

results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_valid = model.predict(X_valid)
    mse = mean_squared_error(y_valid, y_pred_valid)
    r2 = r2_score(y_valid, y_pred_valid)
    results.append({'Model': name, 'MSE': mse, 'R2': r2})

results_df = pd.DataFrame(results)
print(results_df)
```

```python
# entonces el mejor modelo fue el decision tree, ahora vamos a medir su desempeño en el test
best_model = DecisionTreeRegressor(random_state=29)
best_model.fit(X_train, y_train)
y_test_pred = best_model.predict(X_test)
test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)
print(f'Test MSE: {test_mse}, Test R2: {test_r2}')
```

```python
# podemos mejorar el modelo usando grid search para encontrar los mejores hiperparámetros usando validation
from sklearn.model_selection import GridSearchCV
# usaremos un pipeline para escalar los datos y luego aplicar el modelo
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', DecisionTreeRegressor(random_state=29))
])
param_grid = {
    'model__max_depth': [None, 5, 10, 20],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
print(f'Best parameters: {grid_search.best_params_}')  
```

```python
# volvemos el desempeño del mejor modelo con los mejores hiperparámetros en el test 
best_model_tuned = grid_search.best_estimator_
y_test_pred_tuned = best_model_tuned.predict(X_test)
test_mse_tuned = mean_squared_error(y_test, y_test_pred_tuned)
test_r2_tuned = r2_score(y_test, y_test_pred_tuned)
print(f'Tuned Test MSE: {test_mse_tuned}, Tuned Test R2: {test_r2_tuned}')
```

```python
# que variables fueron las mas importantes en el modelo de decision tree pero sin los hiperparámetros
feature_importances = best_model.feature_importances_
importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)
```
