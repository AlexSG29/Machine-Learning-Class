# Diccionario de Variables del Dataset

A continuación se describe el significado de cada variable incluida en el dataset:

---

### **Year**
Año calendario del registro, abarcando desde **1960 hasta 2024**.

---

### **GDP_Current_USD**
Producto Interno Bruto (PIB) expresado en **dólares estadounidenses corrientes**.  
Representa el valor total de los bienes y servicios producidos en el país durante el año, sin ajustes por inflación.

---

### **GDP_Growth_Percent**
Tasa de crecimiento anual del PIB, expresada en **porcentaje (%)**.  
Indica cuánto creció o decreció la economía respecto al año anterior.

---

### **GDP_Per_Capita_USD**
PIB por habitante, expresado en **dólares estadounidenses**.  
Se calcula dividiendo el PIB total entre la población. Mide, de manera aproximada, el nivel promedio de ingreso/economía por persona.

---

### **Inflation_Rate_Percent**
Tasa anual de inflación al consumidor, expresada en **%**.  
Refleja el cambio promedio en los precios de bienes y servicios durante el año.

---

### **Official_Exchange_Rate**
Tipo de cambio oficial, expresado como **riales iraníes (IRR) por 1 dólar estadounidense (USD)**.

---

### **Population_Total**
Población total del país en el año correspondiente.

---

### **Population_Growth_Percent**
Tasa de crecimiento anual de la población, en **%**.  
Mide cuánto aumentó o disminuyó la población respecto al año anterior.

---

### **Exports_USD**
Valor total de las **exportaciones** del país durante el año, expresado en **USD**.

---

### **Imports_USD**
Valor total de las **importaciones**, también expresado en **USD**.

---

### **Trade_Percent_GDP**
Comercio total como porcentaje del PIB.  
Generalmente se calcula como:  
**(Exportaciones + Importaciones) / PIB × 100**

---

### **Industry_Value_Added_Percent_GDP**
Participación del sector industrial en el PIB, expresada en **%**.  
Incluye manufactura, energía, construcción y minería.

---

### **Trade_Balance**
Balanza comercial, calculada como:  
**Exportaciones − Importaciones (en USD)**  
Un valor positivo indica superávit comercial; uno negativo, déficit.

---

```python
# leemos la base de datos
import pandas as pd
df = pd.read_csv('data.csv')
# mostramos las primeras filas
df.head(3)
```

```python
# grafica completa de scatter plot matrix
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df)
plt.show()
```

```python
# leer otra data
df_nuevo = pd.read_csv('fao.csv')
df_nuevo.head()
```

```python
# filtramos cuando la columna item tenga 'Agriculture' aunque sea parte de la palabra
df_agriculture = df_nuevo[df_nuevo['Item'].str.contains('Agriculture', case=False, na=False)]
df_agriculture.head()  
```

```python
# voy a traerme de la base de datos original solo desde 1970 en adelante
df_1970 = df[df['Year'] >= 1970]
df_1970.head()
```

```python
# unimos las dos bases de datos por la columna 'Year' solo me traigo el año y el value
df_agriculture_yearly = df_agriculture.groupby('Year')['Value'].sum().reset_index()
df_merged = pd.merge(df_1970, df_agriculture_yearly, on='Year', how='inner')
df_merged.head()
```

```python
sns.pairplot(df_merged)
plt.show()
```

Bien ahora vamos a predecir el Value de esta nueva base de datos. 

```python
# vamos a llamar todos los modelos de regresion
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# medimos el desempeño de cada modelo
X = df_merged.drop(['Value', 'Unnamed: 0', 'Year'], axis=1)  # variables predictoras
y = df_merged['Value']  # variable objetivo
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=29)
```

```python
# medimos el desempeño de cada modelo
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=29),
    'Random Forest': RandomForestRegressor(random_state=29),
    'Support Vector Regressor': SVR()
}
```

```python
# imprimimos en una tabla el desempeño de cada modelo
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results.append({'Model': name, 'MSE': mse, 'R2 Score': r2})
results_df = pd.DataFrame(results)
results_df.sort_values(by='R2 Score', ascending=False)
```

```python
# vale ahora vamos a usar siempre a partir de ahora el modelo de decision tree que fue el mejor
best_model = DecisionTreeRegressor(random_state=29)
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'MSE: {mse}, R2 Score: {r2}')
```

```python
# vamos a ajustar el modelo de decision tree con los mejores hiperparámetros usando GridSearchCV
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth': [None, 5, 10, 20, 50, 100, 200],
    'min_samples_split': [2, 5, 10, 20, 50, 100],
    'min_samples_leaf': [1, 2, 4, 10, 20, 50],
    'random_state': [29, 42, 100],
    'max_features': [None, 'sqrt', 'log2'],
}
grid_search = GridSearchCV(estimator=best_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=0)
grid_search.fit(X_train, y_train)
print(f'Best Hyperparameters: {grid_search.best_params_}')
```

```python
# imprimir el desempeño del modelo ajustado con los mejores hiperparámetros
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'MSE: {mse}, R2 Score: {r2}')
```

```python
# vamos a usar joblib para guardar el modelo ajustado y luego cargarlo para hacer predicciones
import joblib
joblib.dump(grid_search.best_estimator_, 'mejor_modelo_decision_tree.pkl')
```

```python
# vamos a hacer una predicción y evaluar el modelo ajustado
mejor_modelo = joblib.load('mejor_modelo_decision_tree.pkl')
y_pred = mejor_modelo.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'MSE: {mse}, R2 Score: {r2}')
```

```python
# ahora vamos a ver la importancia de los predictores en el modelo de decision tree
importances = best_model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df.sort_values(by='Importance', ascending=False, inplace=True)
print(feature_importance_df)
```

```python
# veamos en grafica
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance in Decision Tree Regressor')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
# lo mismo usando permutacion importance
from sklearn.inspection import permutation_importance
perm_importance = permutation_importance(best_model, X_test, y_test, n_repeats=10, random_state=29)
perm_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': perm_importance.importances_mean})
perm_importance_df.sort_values(by='Importance', ascending=False, inplace=True)
print(perm_importance_df)   
```

```python
# vemos en grafica
sns.barplot(x='Importance', y='Feature', data=perm_importance_df)
plt.title('Permutation Feature Importance in Decision Tree Regressor')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
