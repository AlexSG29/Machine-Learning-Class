## Prueba de Ydata Profiling

### Prueba con la base de vinos no categorica

```python
import pandas as pd
# Cargamos la data
data_red_wine = pd.read_csv('./wine+quality/winequality-red.csv', sep=';')
data_white_wine = pd.read_csv('./wine+quality/winequality-white.csv', sep=';')
```

```python
# vemos los datos
data_red_wine.head()
```

```python
from ydata_profiling import ProfileReport
# Creamos el reporte
profile_red = ProfileReport(data_red_wine, title="Reporte de Vino Tinto", explorative=True)
```

```python
#vemos el reporte en html
profile_red.to_file("reporte_vino_tinto.html")
```

### Prueba con variables categoricas

```python
# cargamos la nueva data

apt_data = pd.read_csv(
    './apartment_data/apartments_for_rent_classified_10K.csv',
    encoding='cp1252',
    engine='python',        # parser más tolerante
    sep=';',                # ajusta si el separador no es coma
    quotechar='"',          # o cambia si el archivo usa otro
    on_bad_lines='skip'     # o 'warn' para depurar
)

apt_data.head()
```

```python
from ydata_profiling import ProfileReport
# Creamos el reporte
profile_apt = ProfileReport(apt_data, title="Reporte de Apartamentos", explorative=True)
#vemos el reporte en html
profile_apt.to_file("reporte_apartamentos.html")
```
