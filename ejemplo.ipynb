```python
import numpy as np

x = np.arange(-30, 30, 0.5)
x
```

```python
# construimos un nuevo y1 
ruido = np.random.rand(len(x))*40
y1 = 4*x + 6
y2 = y1 + ruido # el del ruido

# el ruido lo que hace es agregarle una variacion aleatoria a cada punto de y1
# esto se usa para simular datos reales, que siempre tienen algun grado de incertidumbre o variabilidad.
```

```python
import matplotlib.pyplot as plt

#vamos a graficas ambas series de datos y1 e y2
#ambos en la misma grafica

plt.scatter(x, y1, color='blue', label='y1 (sin ruido)')
plt.scatter(x, y2, color='red', label='y2 (con ruido)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparaci√≥n de y1 e y2')
plt.legend()
plt.show()
```

```python
# regresion logistica con sigma

sigma  = lambda z: 1 / (1 + np.exp(-z))
z = np.linspace(-10, 10, 100)
plt.plot(z, sigma(z))
plt.title('Funci√≥n Sigmoide')
plt.xlabel('z')
plt.ylabel('œÉ(z)')
plt.grid()
plt.show()
```
